# pandas :: series

### pandas의 구성

### Series

- Column Vector를 표현하는 object

```python
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
```

```python
list_data = [1,2,3,4,5]
example_obj = Series(data = list_data)
example_obj
```

```python
list_data = [1,2,3,4,5]
list_name = ["a","b","c","d","e"]

# index 이름을 지정
example_obj = Series(data = list_data, index=list_name) 
example_obj
```

```python
example_obj.index  # index 리스트만
```

```python
example_obj.values  # 값 리스트만
```

```python
type(example_obj.values)
```

```python
dict_data = {"a":1, "b":2, "c":3, "d":4, "e":5}

# data type, series 이름 설정
example_obj = Series(dict_data, dtype=np.float32, name="example_data")
example_obj
```

```python
# data index에 접근하기
example_obj["a"]
```

```python
# data index에 값 할당하기
example_obj["a"] = 3.2
example_obj
```

```python
example_obj[example_obj > 2]
```

```python
example_obj * 2
```

```python
np.exp(example_obj)
```

```python
np.abs(example_obj) 
```

```python
np.log(example_obj)
```

```python
"b" in example_obj
```
